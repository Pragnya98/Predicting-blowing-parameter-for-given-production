```python
# building model
test_dat = test_data.to_numpy()
X = data.get(['CORR_PRODUCTION'])
Y = data.drop(['CORR_PRODUCTION'], axis = 1)
x1 = X.to_numpy()
y1 = Y.to_numpy()
mod = RF()
mod.fit(x1,y1)
pred = mod.predict(test_data)
pred = pd.DataFrame(pred)
sub = test_data
col = Y.columns
for j in range (len(col)):
    sub[col[j]] = pred.iloc[0:,j]
sub
```

```python
# r-square values for model accuracy
r2 = []
for i in range (1, len(col)+1):
    y_actual = list(data.iloc[0:,i])
    y_predicted = list(sub.iloc[0:,i])
    r= r2_score(y_actual, y_predicted)
    r2.append(r)
newcol = col
r2 = pd.DataFrame(r2)
newcol = pd.DataFrame(newcol)
error = pd.concat([newcol,r2], axis = 1)
error = error.T
error.index = ['Parameters','R-Square Values']
error
```

```python
#making a function for predicting parameters of clusters on the basis of model build above

def MeanminMax(x):
    return pd.Series(index=['mean','min','max'],data=[x.mean(),x.min(),x.max()])
def mod_cluster(z):
    pred1 = mod.predict(z)
    pred1 = pd.DataFrame(pred1)
    sub1 = z
    col = Y.columns
    for j in range (len(col)):
        sub1[col[j]] = pred1.iloc[0:,j]
    return sub1 
def range_cluster(a):
    df = mod_cluster(a)
    return df.apply(MeanminMax)
    
```
